### nested if else

```python
to_minutes = 24 * 60
name_of_units = "minutes"

# use of functions . they are blocks of codes
def days_to_units(num_of_days):
    if num_of_days > 0:
        return f"{num_of_days} days are {num_of_days * to_minutes} {name_of_units}"
    elif num_of_days == 0:
        return "you entered a zero"
    else:
        return "You entered an negative value so no convertion for you"

def validate_and_execute(user_input):
    if user_input.isdigit():
        days = int(user_input)

        if days > 0:
            print(days_to_units(days))
        elif days == 0:
            print("you entered a valid number")
    else:
        print("ypur input is anot a number")

user_input = input("Hey user, enter a number of days and I will convert it to hours!\n")
validate_and_execute(user_input)
```

```python
to_minutes = 24 * 60
name_of_units = "minutes"

# use of functions . they are blocks of codes
def days_to_units(num_of_days):
    if num_of_days > 0:
        return f"{num_of_days} days are {num_of_days * to_minutes} {name_of_units}"
    elif num_of_days == 0:
        return "you entered a zero"
    else:
        return "You entered an negative value so no convertion for you"

def validate_and_execute(user_input):
    try :
        if user_input.isdigit():
            days = int(user_input)

        if days > 0:
            print(days_to_units(days))
        elif days == 0:
            print("you entered a valid number")
    except ValueError:
        print("ypur input is anot a number")

user_input = input("Hey user, enter a number of days and I will convert it to hours!\n")
validate_and_execute(user_input)
```
